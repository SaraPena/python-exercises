```python
# 1. Copy the code from the lesson to create a dataframe full of student grades.

import pandas as pd
import numpy as np

np.random.seed(123)

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

# randomly generate scores for each student for each subject
# note that all the values need to have the same length here
math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))

df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})

type(df)
```

```python
# Create a column named passing_english that indicates whether each student has a 
# passing grade in reading.

# odd that passing english is based on reading grade
# adding column to dataframe, passing grade over 70
df['passing_english'] = df.english > 70
df
```

```python
# Sort the english grades by the passing_english column. How are duplicates handled?

# this syntax works with one column name
df.sort_values(by = 'passing_english')
```

```python
# Sort the english grades first by passing_english and then by student name.
# All the students that are failing english should be first, and within the 
# students that are failing english they should be ordered alphabetically.
# The same should be true for the students passing english. 
# (Hint: you can pass a list to the .sort_values method)

# need sort_values with 2 column names
df.sort_values(by = ['passing_english', 'name'])
```

```python
# Sort the english grades first by passing_english, and then by the actual
# english grade, similar to how we did in the last step.
df.sort_values(by = ['passing_english', 'english'])
```

```python
# Calculate each students overall grade and add it as a column on the dataframe. 
# The overall grade is the average of the math, english, and reading grades.

# mean doesn't work if passing_english column is in place, drop passing_english
# NOTE = only need to do this one time, commenting out when re-running cell
#df.drop(columns=['passing_english'], inplace=True)

# exobrain https://datatofish.com/average-column-row-dataframe/
df['overall_avg'] = df.mean(axis=1)
# verify results
df
```

```python
# Rename the cty column to city.

# use .rename, used = to make permenant change
mpg = mpg.rename(columns = {'cty': 'city'})
# print to screen
mpg
```

```python
# Rename the hwy column to highway.

# use .rename used = to make permenant change
mpg = mpg.rename(columns = {'hwy': 'highway'})
# print to screen
mpg
```

```python
# Do any cars have better city mileage than highway mileage?

# use sort_values to get highest city mpg
# mpg.sort_values(by='city', ascending=False)

# create series for city, hwy, and city_better to compare values
city = mpg.city
hwy = mpg.highway
city_better = hwy[city > hwy]
# create new dataframe to show city, hwy, and city_better values
milage_comp = pd.DataFrame({'city': city, 'hwy': hwy, 'city_better': city_better})
# print new dataframe
print(milage_comp)
```

```python
# Create a column named mileage_difference this column should contain the
# difference between highway and city mileage for each car.

# create series for city, hwy, and mileage_difference
city = mpg.city
hwy = mpg.highway
mileage_difference = hwy - city
# add column 
mpg['mileage_difference'] = mileage_difference
mpg
```

```python
# Which car (or cars) has the highest mileage difference?

# this gets a list with the largest difference shown
mpg.sort_values(by = 'mileage_difference', ascending=False).head(2)

# adding nlargest doesn't seem to work, adding .max() only gets 1 car
```

```python
# Which compact class car has the lowest highway mileage? The best?

# sort dataframe by highway get head for worst and tail for best
worst = mpg.sort_values(by = 'highway').head(5)
print(worst)

best = mpg.sort_values(by = 'highway').tail(2)
print(best)
```

```python
# Create a column named average_mileage that is the mean of the city and highway mileage.

# create series for city, hwy, and mileage_difference
city = mpg.city
hwy = mpg.highway
average_mileage = (hwy + city)/2
# add column 
mpg['average_mileage'] = average_mileage
mpg
```
