```python
import matplotlib.pyplot as plt
from random import randint
import math
```

```python
# 1. Use matplotlib to plot the following equation: y = x^2 - x + 2
# You'll need to write the code that generates the x and y points.
# Add an anotation for the point 0, 0, the origin.

x = list(range(-49, 50))
y = [n**2-n+2 for n in x]
plt.plot(x,y)
plt.annotate('origin', xy=(0, 0), xytext=(-3, 500),
             arrowprops={'arrowstyle': '->'})
```

```python
# 2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):
# y = sqrtx
# y = x^3
# y = 2^x
# You can use functions from the math module to help implement some of the equations above
x1 = list(range(0, 25))
y1 = [n**.5 for n in x1]

plt.plot(x1, y1)
plt.title('$y=\sqrt{x}$')
plt.show()
```

```python
x2 = list(range(0, 25))
y2 = [n**3 for n in x2]

plt.plot(x2, y2)
plt.title('$y=x^3$')
plt.show()
```

```python
x3 = list(range(0, 10))
y3 = [2**n for n in x3]
plt.plot(x3, y3)
plt.title('$y=2^x$')
plt.show()
```

```python
x = list(range(-49, 50))
y = [n**2-n+2 for n in x]
plt.plot(x,y)
plt.title('$y = x^2 - x + 2$')
plt.show()
```

```python
# 3. Combine the figures you created in the last step into one large figure with 4 subplots.
plt.figure(figsize=(16, 9))

plt.subplot(2, 2, 1)
x1 = list(range(0, 25))
y1 = [n**.5 for n in x1]
plt.plot(x1, y1)
plt.title('$y=\sqrt{x}$')


plt.subplot(2, 2, 2)
x2 = list(range(0, 10))
y2 = [n**3 for n in x2]
plt.plot(x2, y2)
plt.title('$y=x^3$')


plt.subplot(2, 2, 3)
x3 = list(range(0, 10))
y3 = [2**n for n in x3]
plt.plot(x3, y3)
plt.title('$y=2^x$')


plt.subplot(2, 2, 4)
x = list(range(-25, 25))
y = [n**2-n+2 for n in x]
plt.plot(x,y)
plt.title('$y = x^2 - x + 2$')

plt.show()
```

```python
# 4. Comine the figures you created in the last step into one figure where each of the 4 equations 
# has a different color for the points. 
# Be sure to include a legend.

plt.figure(figsize=(16, 9))

x1 = list(range(0, 25))
y1 = [n**.5 for n in x1]
plt.plot(x1, y1, label='$y=\sqrt{x}$')

x2 = list(range(0, 10))
y2 = [n**3 for n in x2]
plt.plot(x2, y2, label='$y=x^3$')

x3 = list(range(0, 10))
y3 = [2**n for n in x3]
plt.plot(x3, y3, label='$y=2^x$')

x = list(range(0, 25))
y = [n**2-n+2 for n in x]
plt.plot(x,y,label='$y = x^2 - x + 2$')

plt.legend()
plt.show()
```
