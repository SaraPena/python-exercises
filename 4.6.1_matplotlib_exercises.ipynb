```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
print(plt)
```

```python
x = list(range(-150,150))
y = [n**2-n+2 for n in x]
plt.plot(x,y)
plt.annotate('Origin', xy = (0,0), xytext = (0,5000), arrowprops = {'facecolor': 'blue'})
a = plt.show()
```

```python
x = list(range(1,100))
y = [math.sqrt(n) for n in x]
plt.plot(x, y)
plt.title('$\sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$y$')
b = plt.show()
```

```python
x = list(range(-100,100))
y = [n**3 for n in x]
plt.plot(x,y)
plt.title('$x^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
c = plt.show()
```

```python
x = [-2*math.pi,-1.75*math.pi, -1.5*math.pi,-1.25*math.pi,-math.pi,-.75*math.pi,-.5*math.pi,-.25*math.pi,0,.25*math.pi,.5*math.pi,.75*math.pi, math.pi, 1.25*math.pi, 1.5*math.pi, 1.75*math.pi, 2*math.pi]
y = [math.tan(n) for n in x]
plt.scatter(x,y, c ='blue')
plt.plot(x, y, c = 'blue')
plt.ylim(-5,5)
plt.title('$tan(x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
d = plt.show()
```
