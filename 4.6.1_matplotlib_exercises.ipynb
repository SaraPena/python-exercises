1. Use matplotlib to plot the following equation: 
    
    $y = x^2 - x + 2$
   
   Add an anotation for the poin [0,0], the origin.

```python
x = list(range(-150,150))
y = [n**2-n+2 for n in x]
plt.plot(x,y)
plt.annotate('Origin', xy = (0,0), xytext = (0,5000), arrowprops = {'facecolor': 'blue'})
a = plt.show()
```

3. Combine the figures you created in the last step into one large figure with four subplots.

```python
plt.figure(figsize = (12,8))
plt.suptitle('Graphs of $y = \sqrt{x}$,  $y = x^3$,  $y = tan(x)$,  $y = 2^x$')

plt.subplot(2,2,1)
x = list(range(1,100))
y = [math.sqrt(n) for n in x]
plt.plot(x, y)
plt.title('$\sqrt{x}$')
plt.ylabel('$y$')

plt.subplot(2,2,2)
x = list(range(-100,100))
y = [n**3 for n in x]
plt.plot(x,y)
plt.title('$x^3$')
plt.ylabel('$y$')

plt.subplot(2,2,3)
x = [x*.0002*math.pi for x in range(-10000,10001)]
y = [math.tan(n) for n in x]
plt.scatter(x,y, c ='blue')
plt.plot(x, y, c = 'blue')
plt.ylim(-5,5)
plt.xlim(-2*math.pi, 2*math.pi)
plt.title('$tan(x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.subplot(2,2,4)
x = list(range(-6,7))
y = [2**n for n in x]
plt.plot(x,y)
plt.title('$2^n$')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.show()
```

4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend.

```python
x = range(0,101)
sqrt_x = [math.sqrt(n) for n in x]

a = list(range(-100,100))
b_three = [n**3 for n in a]

c = [c*.0002*math.pi for c in range(-50000,50000)]
tan_d = [math.tan(n) for n in c]

j = list(range(-6,7))
k = [2**n for n in j]

plt.plot(x,sqrt_x, c = 'blue', label = '$\sqrt{x}$')
plt.plot(a, b_three, c = 'red', label = '$x^3$')
plt.plot(c, tan_d, c = 'purple', label = '$tan(x)$')
plt.plot(j, k, c = 'green', label = '$2^n$')

plt.ylim(-20,20)
plt.xlim(-5,5)

plt.legend(loc = 'upper right')
plt.title('Curves of $y = \sqrt{x}$,  $y = x^3$,  $y = tan(x)$,  $y = 2^x$')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.show()
```
