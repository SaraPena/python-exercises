## Function: write_to_json(obj, file_name)

```python
def write_to_json(obj, file_name):
    with open(file_name, 'w') as f:
        json.dump(obj, f, ensure_ascii=False, indent=2)
```

```python
write_to_json(students, 'students.json')
```
