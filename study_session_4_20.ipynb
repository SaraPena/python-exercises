```python
Mammals = data('Mammals')
Mammals
```

```python
len(Mammals[Mammals['hoppers'] == True])
```

```python
speed_col = Mammals['speed']
speed_col
```

```python
median = speed_col.median()
median
```

```python
Mammals[(Mammals['hoppers'] == True) & (Mammals['speed'] > median)]
```

```python
hoppers_median = Mammals[(Mammals['hoppers'] == True) & (Mammals['speed'] > median)]
hoppers_median
```

```python
len_mammals = len(Mammals)
len_mammals
```

```python
len(hoppers_median)/len_mammals
```

```python
(len_specials/len(Mammals['specials']))*100
```
