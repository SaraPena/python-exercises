```python
# Grab departments table
query = """select * from departments"""

departments_df = pd.read_sql(query, get_db_url('employees'))
departments_df
```

```python
# Grab dept_emp table from SQL
query = """select * from dept_emp"""

dept_emp_df = pd.read_sql(query, get_db_url('employees'))
dept_emp_df
```

```python
dept_emp_df.set_index('emp_no',inplace=True)
```

```python
dept_emp_df
```

```python
# Join dept_emp_df and titles_df
title_dept_df = dept_emp_df.join(titles_df, on='emp_no', lsuffix='left_table')
```

```python
title_dept_df
```

```python
# Cleanup - Drop column names
title_dept_df.columns
```

```python
title_dept_df.drop(['from_dateleft_table','to_dateleft_table'], axis=1, inplace=True)
```

```python
title_dept_df
```

```python
# Merge department names to title_dept_df
```

```python
departments_df
```

```python
title_dept_df = title_dept_df.merge(departments_df, how='left')
```

```python
title_dept_df
```

```python
# Now we use groupby to see unique titles by department.
# Create a grouped_df variable/object that groups by dept_name.
grouped_df = title_dept_df.groupby('dept_name')
grouped_df
```

```python
# Use agg method to find unique titles in each dept_name
grouped_df.agg({'title':'nunique'})
```
